```python
import pandas as pd
```

```python
order_items_df = pd.read_csv("./archive/order_items.csv")
order_items_df.info()
```

```python
payments_df = pd.read_csv("./archive/payments.csv")
payments_df.info()
```

```python
customers_df = pd.read_csv("./archive/customers.csv")
customers_df.info()
```

```python
orders_df = pd.read_csv("./archive/orders.csv")
orders_df.info()
```

```python
products_df = pd.read_csv("./archive/products.csv")
products_df.info()
```

```python
# 결측치의 수 확인인
print(order_items_df.isnull().sum(), '\n')
print(payments_df.isnull().sum(), '\n')
print(customers_df.isnull().sum(), '\n')
```

```python
print(orders_df.isnull().sum(), '\n')
print(products_df.isnull().sum(), '\n')
```

```python
# 중복 확인
print(payments_df.duplicated().sum())  # duplicated는 중복이 있는지 보는 함수
print(order_items_df.duplicated().sum())
print(customers_df.duplicated().sum())
print(orders_df.duplicated().sum())
print(products_df.duplicated().sum())
```

```python
# 중복 제거
customers_df = customers_df.drop_duplicates()
customers_df.duplicated().sum()
```

```python
merged = pd.merge(orders_df, payments_df, how='left', on='order_id')
merged = pd.merge(merged, customers_df, how='left', on='customer_id')
merged = pd.merge(merged, order_items_df, how='left', on='order_id')
merged = pd.merge(merged, products_df, how='left', on='product_id')
```

```python
print(merged.duplicated().sum(), '\n')
print(merged.isnull().sum(), '\n')
```

```python
merged = merged.dropna()
merged.isnull().sum()
```

```python
merged.info()
```

```python
merged["product_volume"] = merged["product_length_cm"] * merged["product_height_cm"] * merged["product_width_cm"]
merged.head()
```

```python
# 날짜 관련 열들의 데이터 타입 변경경
merged["order_delivered_timestamp"] = pd.to_datetime(merged["order_delivered_timestamp"])
merged["order_purchase_timestamp"] = pd.to_datetime(merged["order_purchase_timestamp"])
merged["order_approved_at"] = pd.to_datetime(merged["order_approved_at"])
merged["order_estimated_delivery_date"] = pd.to_datetime(merged["order_estimated_delivery_date"])
```

```python
merged.info()
```

```python
# 배달 소요 시간을 나타내는 열 추가
merged["delivered_time"] = merged["order_delivered_timestamp"] - merged["order_purchase_timestamp"]
```

```python
# 숫자형 열만 추출
numeric_columns = merged.select_dtypes(include=['number'])

# IQR 방법을 적용하여 이상치 탐지
for column in numeric_columns:
    Q1 = numeric_columns[column].quantile(0.25)
    Q3 = numeric_columns[column].quantile(0.75)
    
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    
    print(f"Column: {column}")
    print(f"Lower bound: {lower_bound}, Upper bound: {upper_bound}")
    
    # 이상치 필터링
    outliers = merged[(numeric_columns[column] < lower_bound) | (numeric_columns[column] > upper_bound)]
    print("Outliers detected:")
    print(outliers, '\n')
```
